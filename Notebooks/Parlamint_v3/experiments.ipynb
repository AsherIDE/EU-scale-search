```python
# filter de dates die al indexed zijn
def get_remaining_dates(country, dates):
    df_processed_dates_cutout = df_processed_dates[df_processed_dates["key"] == country]
    
    if len(df_processed_dates_cutout) != 0:
        disposable_dates = list(df_processed_dates_cutout["value"])
        remaining_dates = []
        
        for date in dates:
            if date not in disposable_dates:
                remaining_dates.append(date)
                
        return remaining_dates
        
    else:
        return dates

dates = ['2024-12-10' ,'2010-07-07', '2002-01-30', '2005-01-26', '2012-07-05', '2000-01-26', '2006-02-02']
country = "AT"

print(get_remaining_dates(country, dates))
```
